```python
import pandas as pd

# Load datasets
hospital_info = pd.read_csv('Hospital_General_Information.csv')
complications = pd.read_csv('Complications_and_Deaths-Hospital.csv')
unplanned_visits = pd.read_csv('Unplanned_Hospital_Visits-Hospital.csv')

# First look
print(hospital_info.shape)
print(complications.shape)
print(unplanned_visits.shape)
```

```python
# Drop footnote columns
hospital_info = hospital_info.loc[:, ~hospital_info.columns.str.contains('footnote', case=False)]
complications = complications.loc[:, ~complications.columns.str.contains('footnote', case=False)]
unplanned_visits = unplanned_visits.loc[:, ~unplanned_visits.columns.str.contains('footnote', case=False)]
```

```python
# Check for missing values
print(hospital_info.isnull().sum())
print(complications.isnull().sum())
print(unplanned_visits.isnull().sum())
```

```python
#drop the birth resignation since it is not needed
hospital_info = hospital_info.drop(columns=['Meets criteria for birthing friendly designation'])
```

```python
hospital_info.dtypes
```

```python
# Replace 'Not Available' with NaN and convert to numeric
hospital_info['Hospital overall rating'] = pd.to_numeric(
    hospital_info['Hospital overall rating'].replace('Not Available', pd.NA)
)

# Check how many are missing now
print(hospital_info['Hospital overall rating'].isnull().sum())
print(hospital_info['Hospital overall rating'].dtype)
```

```python
print(complications['Score'].unique()[:20])
print(unplanned_visits['Score'].unique()[:20])
```

```python
complications['Score'] = pd.to_numeric(complications['Score'].replace('Not Available', pd.NA))
unplanned_visits['Score'] = pd.to_numeric(unplanned_visits['Score'].replace('Not Available', pd.NA))

# Check missing counts
print(complications['Score'].isnull().sum())
print(unplanned_visits['Score'].isnull().sum())
```

```python
complications['Denominator'] = pd.to_numeric(
    complications['Denominator'].replace({'Not Available': pd.NA, 'Not Applicable': pd.NA})
)
unplanned_visits['Denominator'] = pd.to_numeric(
    unplanned_visits['Denominator'].replace({'Not Available': pd.NA, 'Not Applicable': pd.NA})
)

print(complications['Denominator'].isnull().sum())
print(unplanned_visits['Denominator'].isnull().sum())
```

```python
# Drop rows where hospital rating is missing for our main analysis
hospital_info_clean = hospital_info.dropna(subset=['Hospital overall rating'])

print(f"Hospitals with ratings: {len(hospital_info_clean)}")
print(f"Dropped: {5422 - len(hospital_info_clean)} hospitals")
```

```python
# Merge hospital info with complications and unplanned visits
df = hospital_info_clean.merge(complications, on='Facility ID', how='left')
df = df.merge(unplanned_visits, on='Facility ID', how='left', suffixes=('_comp', '_unplanned'))

print(df.shape)
print(df.head())
```

```python
cols_to_drop = [col for col in df.columns if col.endswith('_x') or col.endswith('_y')]
print(cols_to_drop)
```

```python
# Keep _x versions, drop _y and rename _x to clean names
df = df.drop(columns=[col for col in df.columns if col.endswith('_y')])
df.columns = [col.replace('_x', '') for col in df.columns]

print(df.shape)
print(df.columns.tolist())
```

```python
# Remove duplicate columns
df = df.loc[:, ~df.columns.duplicated()]

print(df.shape)
print(df.columns.tolist())
```

```python
df.to_csv('hospital_clean.csv', index=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.countplot(x='Hospital overall rating', data=df.drop_duplicates('Facility ID'))
plt.title('Distribution of Hospital Star Ratings')
plt.xlabel('Star Rating')
plt.ylabel('Number of Hospitals')
plt.show()
```

```python
avg_rating = df.drop_duplicates('Facility ID').groupby('Hospital Ownership')['Hospital overall rating'].mean().sort_values()

avg_rating.plot(kind='barh', figsize=(10, 6))
plt.title('Average Star Rating by Hospital Ownership Type')
plt.xlabel('Average Star Rating')
plt.tight_layout()
plt.show()
```

```python
avg_by_state = df.drop_duplicates('Facility ID').groupby('State')['Hospital overall rating'].mean().sort_values(ascending=False)

plt.figure(figsize=(20, 6))
avg_by_state.plot(kind='bar')
plt.title('Average Hospital Star Rating by State')
plt.xlabel('State')
plt.ylabel('Average Star Rating')
plt.tight_layout()
plt.show()
```
